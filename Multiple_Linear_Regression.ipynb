```python
## Lets prepare the dataframe
dataset = pd.DataFrame(california.data, columns = california.feature_names)
dataset.head()
```

```python
dataset['Price'] = california.target
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

```python
dataset.head()
```

```python
## Independent and Dependent features
X = dataset.iloc[:,:-1]  ## Independent features
y = dataset.iloc[:, -1]  ## Dependent features


```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 10)
```

```python
X_test.shape, X_train.shape, y_test.shape, y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

## Model Trainig

```python
from sklearn.linear_model import LinearRegression  
regression = LinearRegression() 
```

```python
regression.fit(X_train, y_train)
```

```python
regression
```

```python
## Slope or Coefficient
regression.coef_
```

```python
## Intercepts
regression.intercept_
```

```python
## Prediction for the test data
y_pred = regression.predict(X_test)
```

```python
y_pred
```

```python
## Performance Metrics

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
## R squate and adjusted R square

from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred)
print(score)
```

```python
# display adjusted R-squared
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

## Assumption

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
plt.show()
```

```python
residuals = y_test - y_pred
```

```python
residuals
```

```python
sns.displot(residuals, kind="kde")
plt.show()
```

## Pickling

Python pickle module is used for serialising and de-serialising a Python object structure. Any object in Python can be pickled can be pickled so that it can be saved on disk. What pickle does is that does is that is 'serialises' the object first before writing it to file. Pickling is a way to convert a python object (list, dict, etc) into a character stream. The idea is that this  character stream contains all the information necessary to reconstruct the object in another python script.

```python
import pickle
pickle.dump(regression, open('regressor.pkl', 'wb'))
```

```python
pickle.load(open('regressor.pkl', 'rb'))
```
